```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/population_by_age_group.csv')
df.head()
```

```python
df.info()
df.columns
```

```python
# print(df.columns.tolist())

df_melted = df.melt(id_vars='Country',
                    value_vars=['65+', '25-64 years', '15-24 years', '5-14 years', '0-4 years'],
                    var_name='Age Group',
                    value_name='Population')

# df_melted.tail(25)
```

```python
first_10_countries_df = df[['Country','25-64 years']].head(10)
first_10_countries_df
```

```python
a = input("Enter age group (e.g. 25-64 years): ")

countries = input("Enter 4-5 country names separated by commas: ")
country_list = [c.strip() for c in countries.split(",")]

filtered_df = df[df['Country'].isin(country_list)][['Country', a]]

plt.figure(figsize=(8, 5))
sns.barplot(x=a, y='Country', data=filtered_df)
plt.title(f'Population in Age Group: {a}')
plt.xlabel('Population')
plt.ylabel('Country')
plt.tight_layout()
plt.show()

```
